```python
import pandas as pd
```

```python
#edit commuting_flow data to get number of jobs per census tract
commuting = pd.read_csv('commuting_flow.csv')
```

```python
commuting
```

```python
metrovan_commuting = commuting.copy(deep=True)
```

```python
metrovan_commuting.loc[:,'POW_Code'] = metrovan_commuting['POW_Code'].astype(str)
```

```python
metrovan_commuting = metrovan_commuting.loc[(metrovan_commuting['POW_Code'].str.startswith('933')) & 
                                            (metrovan_commuting['POW_Code'] != '933') &
                                            (metrovan_commuting['Transportation'] == 'Total - Mode of transportation') & 
                                            (metrovan_commuting['POR_Code'] == 0)]
```

```python
metrovan_commuting.loc[:,'number_of_jobs'] = metrovan_commuting.groupby('POW_Code')['Total_Sex'].transform('sum')
```

```python
metrovan_commuting
```

```python
#convert POW_Code to CTUID
metrovan_commuting.loc[:,'POW_Code'] = metrovan_commuting['POW_Code'].apply(lambda x: x[:7] + '.' + x[7:])
```

```python
#total number of jobs in metro van
metrovan_commuting.number_of_jobs.sum()
```

```python
metrovan_commuting['POW_Code'].nunique()
```

```python
metrovan_commuting.to_csv('metrovan_csd_num_jobs.csv')
```

```python
gtfs_pre_morning = pd.read_csv('gtfs_pre_morning_rush.csv')
gtfs_pre_afternoon = pd.read_csv('gtfs_pre_afternoon_rush.csv')
gtfs_post_morning = pd.read_csv('gtfs_post_morning_rush.csv')
gtfs_post_afternoon = pd.read_csv('gtfs_post_afternoon_rush.csv')
```

```python
gtfs_pre_morning = gtfs_pre_morning.set_index('DAUID')
gtfs_pre_afternoon = gtfs_pre_afternoon.set_index('DAUID')
gtfs_post_morning = gtfs_post_morning.set_index('DAUID')
gtfs_post_afternoon = gtfs_post_afternoon.set_index('DAUID')
```

```python
base = gtfs_pre_morning.iloc[:,:26]
```

```python
change1 = gtfs_pre_morning.iloc[:,26:]
change2 = gtfs_pre_afternoon.iloc[:,26:]
change3 = gtfs_post_morning.iloc[:,26:]
change4 = gtfs_post_afternoon.iloc[:,26:]
```

```python
change_morning = change3.subtract(change1)
change_afternoon = change4.subtract(change2)
```

```python
change_morning = base.join(change_morning)
change_afternoon = base.join(change_afternoon)
```

```python
change_morning.to_csv('change_morning.csv')
```
