```python
import pandas as pd
```

```python
pre_8am = pd.read_csv('Service_Area_Pre_8am.csv')
post_8am = pd.read_csv('Service_Area_Post_8am.csv')
```

```python
(pre_8am.iloc[:,:3] == post_8am.iloc[:,:3]).all()
```

```python
pre_8am.rename(columns={'Shape_Area': 'pre_area'}, inplace=True)
post_8am.rename(columns={'Shape_Area': 'post_area'}, inplace=True)
```

```python
change_8am = pd.concat([pre_8am,post_8am.iloc[:,-1:]], axis=1)
```

```python
change_8am.loc[:,'Name'] = change_8am.Name.str.replace('(:).*','')
```

```python
change_8am.loc[:, 'PercChange'] = round(((change_8am.post_area - change_8am.pre_area) / change_8am.pre_area)*100, 2)
```

```python
change_8am = change_8am.loc[:, ['FacilityID', 'Name', 'FromBreak', 'ToBreak', 'PercChange']]
```

```python
pd.set_option('display.max.rows', 100)
```

```python
change_8am['max'] = change_8am.groupby(['Name'])['PercChange'].transform('max')
change_8am = change_8am.loc[change_8am.PercChange > 0].sort_values(['max','Name','PercChange'], ascending=False).drop('max', axis=1)
change_8am.drop(['FacilityID'], axis=1, inplace=True)
```

```python
change_8am['TransitTime'] = change_8am.FromBreak.map(str)+' - '+change_8am.ToBreak.map(str)
```

```python
change_8am = change_8am.loc[:,['Name','TransitTime','PercChange']]
```

```python
change_8am
```

```python
change_8am.to_csv('service_area_8am_change.csv', index=False)
```

```python
pre_4am = pd.read_csv('Service_Area_Pre_4am.csv')
post_4am = pd.read_csv('Service_Area_Post_4am.csv')
pre_4pm = pd.read_csv('Service_Area_Pre_4pm.csv')
post_4pm = pd.read_csv('Service_Area_Post_4pm.csv')
pre_9pm = pd.read_csv('Service_Area_Pre_9pm.csv')
post_9pm = pd.read_csv('Service_Area_Post_9pm.csv')
```

```python
pre_4am.rename(columns={'Shape_Area': 'pre_area'}, inplace=True)
post_4am.rename(columns={'Shape_Area': 'post_area'}, inplace=True)
pre_4pm.rename(columns={'Shape_Area': 'pre_area'}, inplace=True)
post_4pm.rename(columns={'Shape_Area': 'post_area'}, inplace=True)
pre_9pm.rename(columns={'Shape_Area': 'pre_area'}, inplace=True)
post_9pm.rename(columns={'Shape_Area': 'post_area'}, inplace=True)
```

```python
change_4am = pd.concat([pre_4am,post_4am.iloc[:,-1:]], axis=1)
change_4am.loc[:,'Name'] = change_4am.Name.str.replace('(:).*','')
change_4am.loc[:, 'PercChange'] = round(((change_4am.post_area - change_4am.pre_area) / change_4am.pre_area)*100, 2)
change_4am = change_4am.loc[:, ['FacilityID', 'Name', 'FromBreak', 'ToBreak', 'PercChange']]
change_4am['max'] = change_4am.groupby(['Name'])['PercChange'].transform('max')
change_4am = change_4am.loc[change_4am.PercChange > 0].sort_values(['max','Name','PercChange'], ascending=False).drop('max', axis=1)
change_4am.drop(['FacilityID'], axis=1, inplace=True)
change_4am['TransitTime'] = change_4am.FromBreak.map(str)+' - '+change_4am.ToBreak.map(str)
change_4am = change_4am.loc[:,['Name','TransitTime','PercChange']]

change_4pm = pd.concat([pre_4pm,post_4pm.iloc[:,-1:]], axis=1)
change_4pm.loc[:,'Name'] = change_4pm.Name.str.replace('(:).*','')
change_4pm.loc[:, 'PercChange'] = round(((change_4pm.post_area - change_4pm.pre_area) / change_4pm.pre_area)*100, 2)
change_4pm = change_4pm.loc[:, ['FacilityID', 'Name', 'FromBreak', 'ToBreak', 'PercChange']]
change_4pm['max'] = change_4pm.groupby(['Name'])['PercChange'].transform('max')
change_4pm = change_4pm.loc[change_4pm.PercChange > 0].sort_values(['max','Name','PercChange'], ascending=False).drop('max', axis=1)
change_4pm.drop(['FacilityID'], axis=1, inplace=True)
change_4pm['TransitTime'] = change_4pm.FromBreak.map(str)+' - '+change_4pm.ToBreak.map(str)
change_4pm = change_4pm.loc[:,['Name','TransitTime','PercChange']]

change_9pm = pd.concat([pre_9pm,post_9pm.iloc[:,-1:]], axis=1)
change_9pm.loc[:,'Name'] = change_9pm.Name.str.replace('(:).*','')
change_9pm.loc[:, 'PercChange'] = round(((change_9pm.post_area - change_9pm.pre_area) / change_9pm.pre_area)*100, 2)
change_9pm = change_9pm.loc[:, ['FacilityID', 'Name', 'FromBreak', 'ToBreak', 'PercChange']]
change_9pm['max'] = change_9pm.groupby(['Name'])['PercChange'].transform('max')
change_9pm = change_9pm.loc[change_9pm.PercChange > 0].sort_values(['max','Name','PercChange'], ascending=False).drop('max', axis=1)
change_9pm.drop(['FacilityID'], axis=1, inplace=True)
change_9pm['TransitTime'] = change_9pm.FromBreak.map(str)+' - '+change_9pm.ToBreak.map(str)
change_9pm = change_9pm.loc[:,['Name','TransitTime','PercChange']]
```

```python
change_4am.to_csv('service_area_4am_change.csv', index=False)
change_4pm.to_csv('service_area_4pm_change.csv', index=False)
change_9pm.to_csv('service_area_9pm_change.csv', index=False)
```
